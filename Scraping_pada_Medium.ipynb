```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import os
```

```python
urls = {
    'Personal Growth': 'https://medium.com/personal-growth/archive/{0}/{1:02d}/{2:02d}',
    'Mind Cafe': 'https://medium.com/mind-cafe/archive/{0}/{1:02d}/{2:02d}',
    'Elemental': 'https://elemental.medium.com/archive/{0}/{1:02d}/{2:02d}',
    'The Mission': 'https://medium.com/the-mission/archive/{0}/{1:02d}/{2:02d}',
    'P.S. I Love You': 'https://psiloveyou.xyz/archive/{0}/{1:02d}/{2:02d}',
    'Startup Grind': 'https://medium.com/startup-grind/archive/{0}/{1:02d}/{2:02d}'
}
```

```python
def is_leap(year):
    if year % 4 != 0:
        return False
    elif year % 100 != 0:
        return True
    elif year % 400 != 0:
        return False
    else:
        return True
```

```python
def convert_day(day, year):
    month_days = [31, 29 if is_leap(year) else 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    m = 0
    d = 0
    while day > 0:
        d = day
        day -= month_days[m]
        m += 1
    return (m, d)
```

```python
# Memilih tahun dan tanggal acak
year = 2024
selected_days = random.sample([i for i in range(1, 367 if is_leap(year) else 366)], 50)
```

```python
data = []
article_id = 0
i = 0
n = len(selected_days)

# Loop melalui hari-hari yang dipilih dan scraping konten
for d in selected_days:
    i += 1
    month, day = convert_day(d, year)
    date = '{0}-{1:02d}-{2:02d}'.format(year, month, day)
    print(f'{i} / {n} ; {date}')

    for publication, url in urls.items():
        response = requests.get(url.format(year, month, day), allow_redirects=True)
        if not response.url.startswith(url.format(year, month, day)):
            continue
        page = response.content
        soup = BeautifulSoup(page, 'html.parser')

        # Mencari semua artikel
        articles = soup.find_all("div", class_="postArticle postArticle--short js-postArticle js-trackPostPresentation js-trackPostScrolls")
        for article in articles:
            title_tag = article.find("h3", class_="graf--title")
            if title_tag is None:
                continue
            title = title_tag.get_text().strip()

            # Mendapatkan URL artikel
            article_url = article.find_all("a")[3]['href'].split('?')[0]

            # Menambah data ke list
            article_id += 1
            data.append([article_id, article_url, title, publication, date])
```

```python
medium_df = pd.DataFrame(data, columns=['id', 'url', 'title', 'publication', 'date'])
```
